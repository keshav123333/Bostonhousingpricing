```python
import pandas as pd
import urllib.request


url = "http://lib.stat.cmu.edu/datasets/boston"
response = urllib.request.urlopen(url)
raw_data = response.read().decode("utf-8")


lines = raw_data.strip().split('\n')[22:]


data = []
for i in range(0, len(lines)-1, 2):
    line1 = lines[i].strip()
    line2 = lines[i+1].strip()

    if line1 == '' or line2 == '':
        continue

    row = list(map(float, line1.split())) + list(map(float, line2.split()))
    data.append(row)


column_names = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM',
    'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B',
    'LSTAT', 'MEDV'
]

df = pd.DataFrame(data, columns=column_names)

# Final check
print(df.head())
```

```python
import matplotlib.pyplot as plt

def plot_all_vs_target(df, target_col='price'):
    features = df.columns.drop(target_col)  # saare columns except target
    for col in features:
        plt.figure(figsize=(6, 4))
        sns.regplot(x=target_col,y=col,data=df,   color='teal')
        plt.title(f'{col} vs {target_col}')
        plt.xlabel(target_col)
        plt.ylabel(col)
        plt.grid(True)
        plt.tight_layout()
        plt.show()

plot_all_vs_target(df)
```

# Train test split

now we will try to spilt our data on test and train data

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

len(x_train),len(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

# Model training

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.get_params()
```

```python
pred=reg.predict(x_test)
```

```python
plt.scatter(y_test,pred)
```

```python
b=y_test-pred

sns.displot(b,kind="kde")
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(mean_squared_error(y_test,pred))
print(mean_absolute_error(y_test,pred))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,pred)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.scatter(y_test, pred, color='purple', alpha=0.6)
plt.xlabel("Actual MEDV")
plt.ylabel("Predicted MEDV")
plt.title("Actual vs Predicted Prices")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red')  # perfect prediction line
plt.grid(True)
plt.show()
```
